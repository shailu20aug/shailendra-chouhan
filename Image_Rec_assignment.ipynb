### Problem statement: To build a CNN based model which can accurately detect melanoma. Melanoma is a type of cancer that can be deadly if not detected early. It accounts for 75% of skin cancer deaths. A solution which can evaluate images and alert the dermatologists about the presence of melanoma has the potential to reduce a lot of manual effort needed in diagnosis.

#### Class Notes: Import libraries like Tensorflow, Numpy, Pandas, pathlib, glob, keras, matplotlib have been imported below.

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import PIL
import cv2
import pathlib
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')
```

### Important Library Versions:

```python
# Defining the path for train and test images
a= os.getcwd()
b= os.path.join(a,'Image_Assign')
data_dir = pathlib.Path(b)

print (b)
print(data_dir)
```

```python
## Below we will confirm the number of total images in the dataset and their distribution as Train and Test sets.
data_set= list(data_dir.rglob('Picture*.*'))
image_count_set = len(list(data_dir.rglob('Picture*.*')))
print(image_count_set)
print(data_set)
```

```python
## Now we will take a random image from the dataset and check for it's size, format, mode and type. 
img = cv2.imread(data_set[2])
print(img.shape)
image_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
print(image_rgb.shape)
plt.imshow(image_rgb, cmap= plt.cm.gray)
plt.title ("RGB_Picture")

plt.figure()
image_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print(image_gray.shape)

plt.imshow(image_gray, cmap =plt.cm.gray)
plt.title("Gray_picture")
```

```python
# read a image using imread 
input_image = img 
print(input_image.shape)
  
# creating a Histograms Equalization 
# of a image using cv2.equalizeHist() 
equalised_image = cv2.equalizeHist(input_image) 
  
# stacking images side-by-side 
output = np.hstack((input_image, equalised_image)) 
  
# show image input vs output 
cv2.imshow("Images", output) 
  
cv2.waitKey(0) 
cv2.destroyAllWindows() 
```

```python
## Extracting channels
## Blue Channel
image_blue = img
image_blue[:,:,1]=0
image_blue[:,:,2]=0
image_rgb = cv2.cvtColor(image_blue,cv2.COLOR_BGR2RGB)
print(image_rgb.shape)
plt.imshow(image_rgb, cmap= plt.cm.gray)
plt.title ("RGB_Picture")

plt.figure()
image_gray = cv2.cvtColor(image_blue,cv2.COLOR_BGR2GRAY)
print(image_gray.shape)

plt.imshow(image_gray, cmap =plt.cm.gray)
plt.title("Gray_picture")
```

```python
## Green Channel
image_green = img
image_green[:,:,0]=0
image_green[:,:,2]=0
image_rgb = cv2.cvtColor(image_green,cv2.COLOR_BGR2RGB)
print(image_rgb.shape)
plt.imshow(image_rgb, cmap= plt.cm.gray)
plt.title ("RGB_Picture")

plt.figure()
image_gray = cv2.cvtColor(image_green,cv2.COLOR_BGR2GRAY)
print(image_gray.shape)

plt.imshow(image_gray, cmap =plt.cm.gray)
plt.title("Gray_picture")
```

```python
## Red Channel
image_red = img
image_red[:,:,0]=0
image_red[:,:,1]=0
image_rgb = cv2.cvtColor(image_red,cv2.COLOR_BGR2RGB)
print(image_rgb.shape)
plt.imshow(image_rgb, cmap= plt.cm.gray)
plt.title ("RGB_Picture")

plt.figure()
image_gray = cv2.cvtColor(image_red,cv2.COLOR_BGR2GRAY)
print(image_gray.shape)

plt.imshow(image_gray, cmap =plt.cm.gray)
plt.title("Gray_picture")
```
